````python
import numpy as np

def ahp_weight(matrix):
    n = matrix.shape[0]

    #计算几何平均
    geo_mean = np.prod(matrix,axis=1)**(1/n)
    weights = geo_mean / np.sum(geo_mean)

    # 一致性检验
    lambda_max = np.sum(np.dot(matrix, weights) / weights) / n
    CI = (lambda_max - n) / (n - 1)
    RI_dict = {1: 0.0, 2: 0.0, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24, 7: 1.32}
    RI = RI_dict.get(n, 1.49)  # 默认 n=8 时的 RI
    CR = CI / RI if RI != 0 else 0

    return weights, CR

# 示例判断矩阵
A = np.array([
    [1, 3, 5],
    [1/3, 1, 2],
    [1/5, 1/2, 1]
])

weights, CR = ahp_weight(A)
print("权重向量:", weights)
print("一致性比率 CR:", CR)
````

##  AHP 公式推导与一致性检验

在层次分析法中，我们通过 **构造两两比较判断矩阵**，来量化主观判断。设有 $ n $ 个因素，构造判断矩阵 $ A = [a_{ij}]_{n \times n} $，其中：

* $ a_{ij} $ 表示第 $ i $ 个因素相对于第 $ j $ 个因素的重要性。

### 判断矩阵满足：
- $$ a_{ij} > 0 $$
- $$ a_{ij} = \frac{1}{a_{ji}} $$
- $$ a_{ii} = 1 $$
就是对于一个判断矩阵是对角线对称的，对角线（表示同一个因素自我比较）一定是1，同时i对j因素的判断程度是j对i判断的倒数

###  权重计算

采用 **几何平均法** 计算权重向量 $ w = [w_1, w_2, \dots, w_n] $，步骤如下：

1. 计算第 $ i $ 行元素的几何平均值：

$$
GM_i = \left( \prod_{j=1}^{n} a_{ij} \right)^{\frac{1}{n}}
$$

2. 标准化权重向量：

$$
w_i = \frac{GM_i}{\sum_{k=1}^{n} GM_k}
$$

###  一致性检验

判断矩阵不可能完美一致，因此我们需要检验其一致性。首先计算最大特征值 $ \lambda_{\text{max}} $：

$$
\lambda_{\text{max}} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{(A \cdot w)_i}{w_i} \right)
$$

再计算一致性指标（CI）和一致性比率（CR）：

- 一致性指标：

$$
\text{CI} = \frac{\lambda_{\text{max}} - n}{n - 1}
$$

- 一致性比率：

$$
\text{CR} = \frac{\text{CI}}{\text{RI}}
$$

其中 RI 为平均随机一致性指标，取值如下：

| n   | RI   |
|-----|------|
| 1   | 0.00 |
| 2   | 0.00 |
| 3   | 0.58 |
| 4   | 0.90 |
| 5   | 1.12 |
| 6   | 1.24 |
| 7   | 1.32 |

当 **CR < 0.1** 时，判断矩阵被认为具有可接受的一致性。




在本任务中，我们使用 **层次分析法** (AHP) 来计算留学目的国的综合评分。主要包括以下步骤：

## 步骤 1：准则判断矩阵的构造

首先，我们根据不同的准则对备选国家进行评估。假设我们的准则包括 **教育**、**费用**、**生活**、**语言**、**签证**。构造一个准则之间的判断矩阵，每个元素代表两个准则之间的相对重要性。

### 准则判断矩阵：

$$
\begin{bmatrix}
1 & 3 & 4 & 7 & 6 \\
\frac{1}{3} & 1 & 2 & 5 & 4 \\
\frac{1}{4} & \frac{1}{2} & 1 & 3 & 2 \\
\frac{1}{7} & \frac{1}{5} & \frac{1}{3} & 1 & \frac{1}{2} \\
\frac{1}{6} & \frac{1}{4} & \frac{1}{2} & 2 & 1
\end{bmatrix}
$$

每个元素表示一个准则相对于另一个准则的重要性，例如，矩阵中的 $3$ 表示“教育”准则比“费用”准则重要 3 倍。

## 步骤 2：每个准则下的备选方案判断矩阵

接下来，我们为每个准则构造一个成对比较矩阵，用来评估不同备选方案（例如国家）在某一准则下的优劣。每个准则下有一个判断矩阵，其中的元素表示两个备选方案之间的相对重要性。

### 例如，对于“教育”准则的备选方案矩阵：

$$
\begin{bmatrix}
1 & 2 & 3 & 4 \\
\frac{1}{2} & 1 & 1.5 & 2 \\
\frac{1}{3} & \frac{2}{3} & 1 & 1.5 \\
\frac{1}{4} & \frac{1}{2} & \frac{2}{3} & 1
\end{bmatrix}
$$

## 步骤 3：权重计算

对于每个判断矩阵，我们计算 **几何平均数** 来获得每个方案的相对权重。几何平均数计算公式为：

$$
\text{几何平均数} = \left(\prod_{i=1}^n a_{ij}\right)^{1/n}
$$

其中，$a_{ij}$ 是判断矩阵中的元素，$n$ 是矩阵的维度。计算得到的权重经过归一化处理，确保所有权重的和为 1。

### 一致性检验

计算 **最大特征值** $\lambda_{\text{max}}$ 和 **一致性指标** $CI$，公式为：

$$
\lambda_{\text{max}} = \frac{1}{n} \sum_{i=1}^n \frac{a_i w_i}{w_i}
$$

$$
CI = \frac{\lambda_{\text{max}} - n}{n - 1}
$$

根据一致性指标 $CI$，我们可以计算 **一致性比率** $CR$：

$$
CR = \frac{CI}{RI}
$$

其中，$RI$ 是随机一致性指标，取决于矩阵的维度 $n$。如果 $CR$ 小于 0.1，则判断矩阵的一致性可以接受。

## 步骤 4：综合评分计算

通过准则的权重与备选方案的权重进行加权平均，得到每个备选方案的最终综合评分。计算公式为：

$$
\text{综合评分} = \sum_{i=1}^n w_i \cdot a_{ij}
$$

其中，$w_i$ 是准则的权重，$a_{ij}$ 是每个准则下备选方案的权重。

## 步骤 5：结果展示

通过 **Seaborn** 绘制热力图，展示各个准则下备选方案的权重分布。热力图的颜色根据权重的大小进行变化，帮助直观了解每个准则对各个国家的影响。

最终，根据各个国家的综合评分进行排序，得到最优的留学目的国。

### 代码解释

1. **`ahp_weight` 函数**：该函数计算给定判断矩阵的权重，并进行一致性检验，返回计算得到的权重和一致性比率。
2. **`create_judgment_matrix` 函数**：该函数根据输入的成对比较值构造判断矩阵，确保矩阵的对称性。
3. **`alt_weights_matrix`**：计算每个准则下备选方案的权重矩阵，最终进行加权求和得到每个国家的综合评分。

### 最终结果

最终，代码会输出每个国家的综合评分，并根据评分排序，输出最适合的留学目的国。
```


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings

# 忽略所有警告
warnings.filterwarnings("ignore")

# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 支持中文的字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 自定义配色
custom_palette = ['#f9f7f7', '#dbe2ef', '#3f72af', '#112d4e']

def ahp_weight(matrix):
    n = matrix.shape[0]
    geo_mean = np.prod(matrix, axis=1) ** (1/n)  # 计算几何平均数
    weights = geo_mean / np.sum(geo_mean)  # 归一化权重

    # 一致性检验
    lambda_max = np.sum(np.dot(matrix, weights) / weights) / n  # 最大特征值
    CI = (lambda_max - n) / (n - 1)  # 一致性指标
    RI_dict = {1: 0.0, 2: 0.0, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24, 7: 1.32}  # 随机一致性指标
    RI = RI_dict.get(n, 1.49)
    CR = CI / RI if RI != 0 else 0  # 一致性比率
    return weights, CR

# --- 第一步：准则判断矩阵 ---
criteria_names = ['教育', '费用', '生活', '语言', '签证']
criteria_matrix = np.array([
    [1,   3,   4,   7,   6],
    [1/3, 1,   2,   5,   4],
    [1/4, 1/2, 1,   3,   2],
    [1/7, 1/5, 1/3, 1,   1/2],
    [1/6, 1/4, 1/2, 2,   1]
])
criteria_weights, cr_crit = ahp_weight(criteria_matrix)

# --- 第二步：每个准则下的备选方案判断矩阵 ---
alternatives = ['美国', '英国', '加拿大', '德国']

def create_judgment_matrix(vals):
    size = len(vals)
    matrix = np.ones((size, size))  # 初始化矩阵
    for i in range(size):
        for j in range(i+1, size):
            matrix[i, j] = vals[i][j]  # 填充上三角
            matrix[j, i] = 1 / vals[i][j]  # 填充下三角
    return matrix

# 每个准则的备选方案的成对比较输入示例
criteria_alt_matrices = {
    '教育': create_judgment_matrix([
        [1, 2, 3, 4],
        [0, 1, 1.5, 2],
        [0, 0, 1, 1.5],
        [0, 0, 0, 1]
    ]),
    '费用': create_judgment_matrix([
        [1, 1/2, 1/3, 1/4],
        [0, 1, 1/2, 1/3],
        [0, 0, 1, 1/2],
        [0, 0, 0, 1]
    ]),
    '生活': create_judgment_matrix([
        [1, 1.5, 1.2, 1.8],
        [0, 1, 1.1, 1.5],
        [0, 0, 1, 1.3],
        [0, 0, 0, 1]
    ]),
    '语言': create_judgment_matrix([
        [1, 1.1, 1.2, 2],
        [0, 1, 1.1, 1.8],
        [0, 0, 1, 1.6],
        [0, 0, 0, 1]
    ]),
    '签证': create_judgment_matrix([
        [1, 1.2, 1.5, 2],
        [0, 1, 1.2, 1.7],
        [0, 0, 1, 1.4],
        [0, 0, 0, 1]
    ])
}

# 第三步：计算每个准则下备选方案的权重
alt_weights = []
for key in criteria_names:
    w, cr = ahp_weight(criteria_alt_matrices[key])
    alt_weights.append(w)

# 转换为矩阵：形状 (准则数量, 备选方案数量)
alt_weights_matrix = np.array(alt_weights)

# 第四步：准则权重 × 备选方案权重
final_scores = np.dot(criteria_weights, alt_weights_matrix)

# ---  Seaborn 热力图 ---
df_heat = pd.DataFrame(alt_weights_matrix, index=criteria_names, columns=alternatives)

plt.figure(figsize=(10, 6))
sns.heatmap(df_heat, annot=True, cmap=sns.color_palette(custom_palette), fmt=".2f")
plt.title("各准则下留学国家权重分布 (AHP)", fontsize=18)
plt.xlabel("国家", fontsize=14)
plt.ylabel("评估准则", fontsize=14)
plt.tight_layout()
plt.show()

# 最终结果
final_df = pd.DataFrame({
    '国家': alternatives,
    '综合评分': final_scores
}).sort_values(by='综合评分', ascending=False)

print("🏆 综合得分结果：")
print(final_df)
````
